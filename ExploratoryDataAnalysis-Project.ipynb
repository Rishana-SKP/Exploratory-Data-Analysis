### <font color='brown'>1.Number of Employees in each Team and the percentage splitting with respect to the total employees.</font>

```python
#Number of employees in each team
Count=df['Team'].value_counts()
Count
```

```python
#Percentage splitting
Count=df['Team'].value_counts()
Total_employes=len(df)
percentage=(Count/Total_employes)*100
percentage
```

```python
#Visual Representation in the bar chart
plt.bar(Count.index,Count.values,color='green',width=0.5)
# Customize the plot
plt.xlabel('Team')
plt.ylabel('Percentage')
plt.title('Percentage Splitting of Employees in each Team')
plt.xticks(rotation='vertical')
plt.show()
```

```python
#plot Pie chart
Count.plot(kind='pie')

#Add Title
plt.title('Percentage Split by Team')
plt.show()
```

<b>Upon reviewing the bar chart & pie chart, it's evident that the New Orleans Pelicans have a notably higher percentage split compared to other team members. Additionally, 23 teams exhibit an identical percentage split in the analysis.</b>

### <font color='brown'>4.Find out under which team and position, spending in terms of salary is high.</font>

```python
High_Salary=df.groupby(['Team','Position'])['Salary'].sum()
High_Salary
```

<b>df.groupby(['Team','Position'])['Salary'].sum() is performing a group-by operation on a DataFrame. It's grouping the data by two columns: 'Team' and 'Position'. Then, it selects the 'Salary' column and computes the sum of salaries for each unique combination of 'Team' and 'Position'. Essentially, it calculates the total salary for each team-position pair.

High_Salaye stores the result of the above operation. It likely contains a Pandas Series or DataFrame with the total salaries for each unique team-position combinati</b>on

```python
team_highest_salary=High_Salary.idxmax()
team_highest_salary
```

<b> idxmax() function is used to returns the index (or row label) where the maximum value occurs in the High_Salary Series or DataFrame.Hence it finds the team-position combination that has the highest total salary.</b>

```python
highest_salary_amount=High_Salary.max()
highest_salary_amount
```

<b>max() function retrieves the maximum value from the High_Salary Series or DataFrame. Hence it finds the highest total salary among all team-position combinations.</b>

### <font color='brown'>5.Find if there is any correlation between age and salary , represent it visually.</font>

```python
#Correlation Between Age and Salary
corrltn=df[['Age','Salary']].corr()
corrltn
```

```python
#Plot the correlation heatmap
sns.heatmap(corrltn,annot=True,cmap='coolwarm',fmt=".2f")

#Add title
plt.title('Correlation Between Age & Salary')
plt.show()
```

<b>This correlation matrix offers a comprehensive view of the relationship between 'Age' and 'Salary', revealing the extent of their linear association; a correlation value near 1 signifies a strong positive linear correlation between these variables.|</b>
